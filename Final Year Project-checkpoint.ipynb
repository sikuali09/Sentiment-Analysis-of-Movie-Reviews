```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words("english"))
example = "barnold schwarzenegger has been an icon for action enthusiasts , since the late 80's , but lately his films have been very sloppy and the one-liners are getting worse . \nit's hard seeing arnold as mr . freeze in batman and robin , especially when he says tons of ice jokes , but hey he got 15 million , what's it matter to him ? \nonce again arnold has signed to do another expensive blockbuster"
words = word_tokenize(example)
filtered_sentence = []
for w in words:
    if w not in stop_words:
        filtered_sentence.append(w)
print(filtered_sentence)        
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
cv = CountVectorizer(tokenizer=nltk.word_tokenize , stop_words='english')
movie_train_cv = cv.fit_transform(movie_train.data)
```

```python
cv.vocabulary_ # it converts into dictionary and tell how many times the word comes
```

```python
movie_train_cv.toarray()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf = TfidfTransformer()
movie_train_tfidf = tfidf.fit_transform(movie_train_cv)
```

```python
movie_train_tfidf.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.pipeline import Pipeline
movie_train_clf = Pipeline([('vect',CountVectorizer(stop_words='english')),('tfidf',TfidfTransformer()),('clf',MultinomialNB(fit_prior = False))])
movie_train_clf = movie_train_clf.fit(movie_train.data , movie_train.target)
```

```python
movie_train_clf
```

```python
import numpy as np
predict = movie_train_clf.predict(movie_train.data)
np.mean(predict==movie_train.target)
```

```python
target_names = ['neg' , 'pos']
print(target_names)
```

```python
from sklearn.metrics import classification_report
print(classification_report(movie_train.target, predict, target_names=target_names))
```

```python
from sklearn.svm import SVC
```

```python
movie_train_clf = Pipeline([('vect',CountVectorizer(stop_words='english')),('tfidf',TfidfTransformer()),('clas',SVC())])

movie_train_clas = movie_train_clf.fit(movie_train.data , movie_train.target)
```

```python
predict2 = movie_train_clas.predict(movie_train.data)
np.mean(predict2==movie_train.target)
```

```python
from sklearn.metrics import classification_report
print(classification_report(movie_train.target, predict2, target_names=target_names))
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(movie_train.target , predict2)
```
